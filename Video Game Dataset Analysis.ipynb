```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Reading in Dataset

```python
data= pd.read_csv('C:/Users/jovan/Downloads/archive/vgsales.csv')
```

```python
data.head(20)
```

```python
data.tail()
```

#### Filtering Data

```python
data2=data.loc[data["Publisher"]=='Nintendo']
data2=data2.loc[data2['Year']>=2015]
data2=data2.loc[data2['Platform']=='3DS']
```

```python
data2.groupby(['Name']).sum()
```

#### Graphing Nintendo 3DS Published Games from 2015-2016

```python
f, ax = plt.subplots(figsize=(18,5))
names=[name for name, df in data2.groupby(['Name'])]
y=data2.groupby(['Name']).sum()
plt.bar(names,y['Global_Sales'],color='#2ec77f')
plt.xticks(rotation='vertical',fontsize=10)
plt.ylabel('Millions USD')
plt.title('Global Sales of Nintendo Published 3DS Titles (2015-2016)')
plt.show()
```

#### Filtering data by Genre

```python
data.groupby(['Genre'])['Global_Sales'].sum().sort_values(ascending=False)
```

#### Graphing Sales by Genre

```python
genres=[genre for genre, df in data.groupby(['Genre'])]
y=data.groupby(['Genre']).sum()
plt.bar(genres,y['Global_Sales'],color='#fa0f90')
plt.xticks(rotation='vertical')
plt.ylabel('Millions USD')
plt.title('Global Sales by Video Game Genre')
plt.show()
```

#### Filtering by Publisher

```python
data4=data.groupby(['Publisher']).sum()
data4=data4.loc[data4['Global_Sales']>250]
```

```python
data4
```

#### Bar chart for top 10 Video Game Publishers

```python
publishers=[publisher for publisher, df in data4.groupby(['Publisher'])]
y=data4.groupby(['Publisher']).sum()
plt.bar(publishers,y['Global_Sales'],color='#b207f5')
plt.xticks(rotation='vertical')
plt.ylabel('Millions USD')
plt.title('Top 10 Highest Selling Video Game Publishers')
plt.show()
```

#### Pie chart for Video Game Publishers

```python
plt.pie(y['Global_Sales'],labels=publishers,autopct='%1.1f%%',shadow=True,)
plt.axis('equal')
plt.title('Top 10 Highest Selling Video Game Publishers')
plt.show()
```
