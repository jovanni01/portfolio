#### Restructuring Data to have col=Regions

```python

graph_df=pd.DataFrame()

for region in avocado['region'].unique()[:13]:    
    print(region)
    region_df=avocado.copy()[avocado['region']==region]
    region_df.set_index('Date',inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_25MA']=region_df['AveragePrice'].rolling(25).mean()
    if graph_df.empty: 
        graph_df=region_df[[f'{region}_25MA']]
    else:
        graph_df=graph_df.join(region_df[f'{region}_25MA'])
graph_df.tail()
```
